# **2-1 Fibbonaci Number**

**Python**

```python
n = int(input())

if n<=1:
    print(n)  
    quit()

def fibo(n):
    a, b = 0, 1
    for _ in range(n-1):
        c = a + b
        b, a = c, b
    print(c)

fibo(n)
```

**C++**

```python
%%writefile fibonacci.cpp
#include <iostream>
#include <cassert>
#include <vector>
using namespace std;

int fibonacci_naive(int n) {
    if (n <= 1)
        return n;

    return fibonacci_naive(n - 1) + fibonacci_naive(n - 2);
}

long long fibonacci_fast(int n) {
    // write your code here
    vector<long long> fibo(n+1,0);
    fibo[0] = 0;
    fibo[1] = 1;
    for(int i = 2; i <= n; i++){
        fibo[i] = fibo[i-1] + fibo[i-2];
    }
    return fibo[n];
}

void test_solution() {
    assert(fibonacci_fast(3) == 2);
    assert(fibonacci_fast(10) == 55);
    for (int n = 0; n < 20; ++n)
        assert(fibonacci_fast(n) == fibonacci_naive(n));
}

int main() {
    int n = 0;
    std::cin >> n;

    // std::cout << fibonacci_naive(n) << '\n';
    // test_solution();
    std::cout << fibonacci_fast(n) << '\n';
    return 0;
}
```

```python
!g++ fibonacci.cpp -o fibonacci
!./fibonacci
```

# **2-2 Last digit of a large Fibbonaci**

**Python**

```python
n = int(input())
if n<=1:
    print(n)
def lastdigitfibo(n):
    a, b = 0, 1
    for _ in range(n-1):
        c = a + b
        c = c%10
        b, a = c, b
    print(c)

lastdigitfibo(n)
```

**C++**

```python
%%writefile fibonacci_last_digit.cpp
#include <iostream>
#include <vector>
using namespace std;

int get_fibonacci_last_digit_naive(int n) {
    if (n <= 1)
        return n;

    int previous = 0;
    int current  = 1;

    for (int i = 0; i < n - 1; ++i) {
        int tmp_previous = previous;
        previous = current;
        current = tmp_previous + current;
    }

    return current % 10;
}

int get_fibonacci_last_digit_fast(int n){
    vector<int> fibo(n+1,0);
    fibo[0] = 0;
    fibo[1] = 1;
    for(int i = 2; i <= n; i++){
        fibo[i] = (fibo[i-1] + fibo[i-2])%10;
    }
    return fibo[n];
}

int main() {
    int n;
    std::cin >> n;
    // int c = get_fibonacci_last_digit_naive(n);
    int c = get_fibonacci_last_digit_fast(n);
    std::cout << c << '\n';
}
```

# **2-3 Greatest Common Divisor**

**Python**

```python
a, b = [int(i) for i in input().split()]
    
def computeGCD(a, b): 
    while(b):
        a, b = b, a % b 
        
    return a 
  
print (computeGCD(a,b)) 
```

**C++**

```python
%%writefile gcd.cpp
#include <iostream>

int gcd_naive(int a, int b) {
  int current_gcd = 1;
  for (int d = 2; d <= a && d <= b; d++) {
    if (a % d == 0 && b % d == 0) {
      if (d > current_gcd) {
        current_gcd = d;
      }
    }
  }
  return current_gcd;
}

int gcd_fast(int a, int b){
  if(b == 0) return a;
  return gcd_fast(b,a%b);
}

int main() {
  int a, b;
  std::cin >> a >> b;
  std::cout << gcd_fast(a, b) << std::endl;
  return 0;
}
```

```python
!g++ gcd.cpp -o gcd
!./gcd
```

# **2-4 Least Common Multiple**

**Python**

```python
a, b = [int(i) for i in input().split()]

def euclid_gcd(a, b):
    if b == 0:
        return a
    c = a%b
    return euclid_gcd(b, c)

if a>b:
    gcd = euclid_gcd(a, b)
else:
    gcd = euclid_gcd(b, a)

print(a*b//gcd)
```

**C++**

```python
%%writefile lcm.cpp
#include <iostream>

long long lcm_naive(int a, int b) {
  for (long l = 1; l <= (long long) a * b; ++l)
    if (l % a == 0 && l % b == 0)
      return l;

  return (long long) a * b;
}

int gcd_fast(int a, int b){
  if(b == 0) return a;
  return gcd_fast(b,a%b);
}

long long lcm_fast(long long a, long long b){
  long long gcd = gcd_fast(a,b);
  return (a*b)/gcd;
}

int main() {
  long long a, b;
  std::cin >> a >> b;
  // std::cout << lcm_naive(a, b) << std::endl;
  std::cout << lcm_fast(a, b) << std::endl;
  return 0;
}
```

```python
!g++ lcm.cpp -o lcm
!./lcm
```
